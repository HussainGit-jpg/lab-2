# Sets

```python
num_2 = {1,2,3,4,5,6}
num_3={7,8,9}
print(num_2.union(num_3))
print(num_2.intersection(num_3))
print(num_2.difference(num_3))
```

# Dictionary

```python
student = {"Ali": 85, "Ahmad": 89, "hussain": 67, "abid": 90}
print("keys:", student.keys())
print("values:", student.values())

#update value
student["Ali"] = 93
print("Update Dictionary: ", student)

#delete a key values pair
del student["Ahmad"]
print("After Deletion: ", student)
```
